```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def approche_gloutonne(mat_N, mat_NE, mat_E):
    x,y = (0,0)
    m = len(mat_N)
    n = len(mat_N[0])
    
    chemin = [(x,y)]
    cout_chemin = 0
    
    while((x,y) != (m-1,n-1) ):
        if(x == m-1) :cost_N = float('inf')
        else: cost_N = mat_N[x][y]
            
        if(y == n-1) : cost_E = float('inf')
        else: cost_E = mat_E[x][y]
        
        if(y == n-1 and x == m-1) : cost_NE = float('inf')
        else : cost_NE = mat_NE[x][y]
        
        min_dir = min(cost_N, cost_NE, cost_E)
        
        if(min_dir == cost_N):
            x += 1
            chemin.append((x,y))
            cout_chemin += min_dir
            continue
        
        if(min_dir == cost_E):
            y += 1
            chemin.append((x,y))
            cout_chemin += min_dir
            continue
        
        if(min_dir == cost_NE):
            x += 1
            y += 1
            chemin.append((x,y))
            cout_chemin += min_dir
            continue
        
    return cout_chemin, chemin
```

```python
def approche_optimale(mat_N, mat_NE, mat_E) :
    m = len(mat_N)
    n = len(mat_N[0])
    
    M = [ [float("inf") for c in range(n)] for l in range(m)]

    M[0][0] = 0
    for x in range(1,m) : M[x][0] = M[x-1][0] + mat_N[x-1][0]
    for y in range(1,n) : M[0][y] = M[0][y-1] + mat_E[0][y-1]

    for x in range(1,m) :
        for y in range(1,n) :
                
            if(x-1 == m-1 and y-1 == n-1):NE = float("inf")
            else:
                NE = mat_NE[x-1][y-1]
                
                if(y-1 == n-1): E = float('inf')
                else: E = mat_E[x][y-1]

                if(x-1 == m-1): N = float('inf')
                else: N = mat_N[x-1][y]
            
            
            M[x][y] = min(
            N + M[x-1][y],
            E + M[x][y-1],
            NE + M[x-1][y-1]
            )
    return M

def afficher(M) :
    m,n = len(M),len(M[0])
    for x in range(m-1,-1,-1) :
        for y in range(n) :
            print(str(M[x][y]) + "\t", end="")
        print("")
        
def accm(x,y,M,mat_N,mat_NE, mat_E) :
    m = len(mat_N)
    n = len(mat_N[0])
    
    if (x,y) == (0,0) :
        print( str((x,y)), end=" " ) 
        return
    if y == 0 :
        accm(x-1,0,M,mat_N,mat_NE, mat_E) 
        c = mat_N[x-1][0]
        print("--" + str(c) + "--> " + str((x,y)), end=" " )
        return
    if x == 0 :
        accm(0,y-1,M,mat_N,mat_NE, mat_E)
        c = mat_E[0][y-1]
        print( "--" + str(c) + "--> " + str((x,y)), end=" " )
        return

    cminN = mat_N[x-1][y] + M[x-1][y]
    cminE = mat_E[x][y-1] + M[x][y-1]
    cminNE = mat_NE[x-1][y-1] + M[x-1][y-1] 
    if cminN <= min(cminE,cminNE) : 
        accm(x-1,y,M,mat_N,mat_NE, mat_E) 
        print("--" + str(mat_N[x-1][y]) + "--> " + str((x,y)), end=" " )
        return
    elif cminE <= cminNE : 
        accm(x,y-1,M,mat_N,mat_NE, mat_E) 
        print("--" + str(mat_E[x][y-1]) + "--> " + str((x,y)), end=" " )
        return
    else :
        accm(x-1,y-1,M,mat_N,mat_NE, mat_E) 
        print( "--" + str(mat_NE[x-1][y-1]) + "--> " + str((x,y)), end=" " )
        return
```

```python
## Main ##

nb_tests = 500
list_val_opti = []
list_val_gloutonne = []
list_dist_relativ = []

for x in range(nb_tests):
    n = random.randint(5,20)
    m = random.randint(5,20)
    mat_N = [[random.randint(0,101) for j in range(n)] for i in range(m)]
    mat_E = [[random.randint(0,101) for j in range(n)] for i in range(m)]
    mat_NE = [[random.randint(0,101) for j in range(n)] for i in range(m)]
    
    M = approche_optimale(mat_N, mat_NE, mat_E)
    cout_glouton, chemin = approche_gloutonne(mat_N, mat_NE, mat_E)
    cout_optimal = M[m-1][n-1]
    
    list_val_gloutonne.append(cout_glouton)
    list_val_opti.append(cout_optimal)
    list_dist_relativ.append(abs((cout_optimal - cout_glouton)/cout_optimal) )
    
    """
    print("Coût optimal de (0,0) à " + str((m-1,n-1)) + " = " + str( M[m-1][n-1] ))
    accm(m-1,n-1,M,mat_N,mat_NE, mat_E)
 
    print("\n\nCoût glouton de (0,0) à " + str((m-1,n-1)) + " = " + str( cout_glouton ))
    print(str(chemin[0]) + " ",end="")
    for elem in chemin[1:]:
        print("--> " + str(elem) + " ", end="")
    """

num_bins = len(list_dist_relativ)//2
plt.hist(list_dist_relativ, num_bins, facecolor = 'blue', alpha = 1)
plt.title("Distance relative entre méthode gloutonne et optimale par valeur")
plt.show()

##########
```
